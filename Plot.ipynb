```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.style.use('science')
plt.rcParams.update({
    "font.family": "serif",   # specify font family here
    "font.serif": ["Arial"],  # specify font here
    "font.size":20})          # specify font size here
```

```python
df = pd.read_csv("info.csv")
df_loss_acc = pd.read_csv("loss_acc.csv")
```

```python
plt.figure(figsize=(8,5))
plt.plot(df_loss_acc["tr_acc"], label="tr acc")
plt.plot(df_loss_acc["va_acc"], label="va acc")
plt.legend()
plt.savefig("tr_va_acc.png")
plt.show()
```
